### Zadanie 3: Rejestrowanie dostepu
Stworz klase `Uzytkownik`. Klasa powinna zawierac atrybuty `imie` i `wiek`. Opracuj deskryptor, ktory bedzie rejestrowal dostep do tych atrybutow za pomoca logowania. Deskryptor powinien logowac informacje o odczycie (`__get__`) oraz zapisie (`__set__`) wartosci atrybutu.

Wzorujemy sie na DEMO `EmailValidator` z notebooka glownego - uzywamy `__set_name__` zeby deskryptor poznal swoja nazwe i `instance.__dict__` jako miejsce skladowania wartosci.

```python
import logging

logging.basicConfig(level=logging.INFO)


class AccessLogger:
    def __set_name__(self, owner, name):
        self.name = name

    def __get__(self, instance, owner):
        if instance is None:
            return self
        value = instance.__dict__.get(self.name)
        logging.info(f"Odczyt atrybutu '{self.name}' = {value!r}")
        return value

    def __set__(self, instance, value):
        logging.info(f"Zapis atrybutu '{self.name}' = {value!r}")
        instance.__dict__[self.name] = value
```

```python
class Uzytkownik:
    imie = AccessLogger()
    wiek = AccessLogger()

    def __init__(self, imie, wiek):
        self.imie = imie
        self.wiek = wiek
```

```python
# Test
u = Uzytkownik("Anna", 25)
print("---")
print("Pobranie imienia:", u.imie)
print("Pobranie wieku:  ", u.wiek)
print("---")
u.wiek = 26
print("Po zmianie wiek:", u.wiek)
```
