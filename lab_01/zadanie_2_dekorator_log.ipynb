### Zadanie 2: Logowanie do pliku
Stworz dekorator, ktory bedzie zapisywal w pliku `*.log` nazwe funkcji dekorowanej, date oraz dlugosc wykonania. Nazwa pliku bedzie podana jako argument dekoratora.

**Protip:** uzyj biblioteki `datetime`. Pamietaj o tym, zeby dekorator zachowal metadane funkcji dekorowanej.

```python
import functools
from datetime import datetime
import time


def logger(filename):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            start = time.perf_counter()
            result = func(*args, **kwargs)
            duration = time.perf_counter() - start
            with open(filename, "a", encoding="utf-8") as f:
                f.write(
                    f"{datetime.now().isoformat()} | "
                    f"{func.__name__} | "
                    f"{duration:.6f} s\n"
                )
            return result
        return wrapper
    return decorator
```

```python
@logger("wykonania.log")
def example_task(n):
    total = 0
    for i in range(n):
        total += i
    return total


@logger("wykonania.log")
def silnia(n):
    wynik = 1
    for i in range(1, n + 1):
        wynik *= i
    return wynik


example_task(1_000_000)
silnia(20)
example_task(500_000)
print("Wykonano. Sprawdz plik wykonania.log")
```

```python
# Podglad logu
with open("wykonania.log", encoding="utf-8") as f:
    print(f.read())
```
