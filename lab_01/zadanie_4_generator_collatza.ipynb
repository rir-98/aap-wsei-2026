### Zadanie 4: Generator ciagu Collatza
Opracuj generator ciagu Collatza. Dla liczby naturalnej n:
- jesli n jest parzyste, dziel przez 2
- jesli n jest nieparzyste, pomnoz przez 3 i dodaj 1

zaczynajac od okreslonej liczby poczatkowej, az do osiagniecia wartosci 1.

```python
def collatz_generator(n):
    yield n
    while n != 1:
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        yield n
```

```python
# Test
for status in collatz_generator(10):
    print(status)
```
