### Zadanie 1: Liczba elementow listy
Stworz dekorator, ktory bedzie odpowiedzialny za wyswietlanie liczby elementow listy, jesli jakakolwiek lista pojawi sie w parametrach funkcji dekorowanej.

**Protip:** uzyj `isinstance` do sprawdzenia czy parametr jest lista. Pamietaj o zachowaniu metadanych funkcji (`functools.wraps`).

```python
import functools


def show_list_length(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        for i, arg in enumerate(args):
            if isinstance(arg, list):
                print(f"args[{i}] -> lista o dlugosci {len(arg)}")
        for name, value in kwargs.items():
            if isinstance(value, list):
                print(f"kwargs['{name}'] -> lista o dlugosci {len(value)}")
        return func(*args, **kwargs)
    return wrapper
```

```python
@show_list_length
def process_data(data_list, name):
    print(f"Przetwarzanie {name}")
    return len(data_list)


process_data([1, 2, 3, 4, 5], "test_a")
process_data([10, 20], "test_b")
process_data("bez listy", "test_c")
```
