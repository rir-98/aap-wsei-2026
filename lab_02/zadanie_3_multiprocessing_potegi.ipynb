### Zadanie 3 (Multiprocessing)
Napisz program, ktory zrownolegli obliczanie sumy kolejnych stu poteg dla kazdej liczby z ciagu liczb naturalnych w duzym zakresie (np. 1 - 10 000).
Uzyj modulu `multiprocessing` oraz gotowej funkcji `calculate_power_sum(n)` z pliku `lab2_functions.py`.

Pamietaj o bezpiecznym uruchamianiu procesow na macOS (`if __name__ == "__main__":`).

```python
import multiprocessing
import time
from lab2_functions import calculate_power_sum

LIMIT = 10_000
```

```python
# 1. Sekwencyjnie - referencja
if __name__ == "__main__":
    start = time.time()
    wyniki_seq = [calculate_power_sum(n) for n in range(1, LIMIT + 1)]
    czas_seq = time.time() - start
    print(f"Sekwencyjnie ({LIMIT} liczb): {czas_seq:.2f} s")
```

```python
# 2. Multiprocessing
if __name__ == "__main__":
    cores = multiprocessing.cpu_count()
    print(f"Praca na {cores} procesach (rdzeniach)...")
    start = time.time()

    with multiprocessing.Pool(processes=cores) as pool:
        wyniki_par = pool.map(calculate_power_sum, range(1, LIMIT + 1))

    czas_par = time.time() - start
    print(f"Rownolegle ({cores} procesy): {czas_par:.2f} s")
    print(f"Przyspieszenie: {czas_seq / czas_par:.1f}x")
```

```python
# 3. Sprawdzenie poprawnosci
if __name__ == "__main__":
    assert wyniki_seq == wyniki_par, "Wyniki sa rozne!"
    print(f"Wyniki sa identyczne. Liczba wynikow: {len(wyniki_par)}")
    print(f"Suma stu poteg dla n=5: {wyniki_par[4]}")
```
