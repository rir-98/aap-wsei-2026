### Zadanie 1 (Threading)
Przy uzyciu publicznego API **Cat Facts** (`https://catfact.ninja/fact`), ktore zwraca przy kazdym wywolaniu losowy fakt na temat kotow:
1. Pobierz sekwencyjnie 20 faktow i zmierz czas calkowitego dzialania programu.
2. Zmodyfikuj kod, aby wysylac zapytania wielowatkowo przy uzyciu `ThreadPoolExecutor`.
3. Porownaj czasy wykonania.

*Podpowiedz: Uzyj `requests.get(URL).json().get('fact')`*

```python
import requests
import time
import concurrent.futures

CAT_API_URL = "https://catfact.ninja/fact"
N_REQUESTS = 20


def fetch_fact(_=None):
    return requests.get(CAT_API_URL).json().get('fact')
```

```python
# 1. Sekwencyjnie
start = time.time()
fakty_seq = [fetch_fact() for _ in range(N_REQUESTS)]
czas_seq = time.time() - start
print(f"Pobrano sekwencyjnie {len(fakty_seq)} faktow w {czas_seq:.2f} s")
```

```python
# 2. Wielowatkowo (ThreadPoolExecutor)
start = time.time()
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    fakty_par = list(executor.map(fetch_fact, range(N_REQUESTS)))
czas_par = time.time() - start
print(f"Pobrano wielowatkowo {len(fakty_par)} faktow w {czas_par:.2f} s")
```

```python
# 3. Porownanie
print(f"\nSekwencyjnie:   {czas_seq:.2f} s")
print(f"Wielowatkowo:   {czas_par:.2f} s")
print(f"Przyspieszenie: {czas_seq / czas_par:.1f}x")

print("\nKilka pierwszych faktow:")
for i, fakt in enumerate(fakty_par[:5], 1):
    print(f"{i}. {fakt}")
```
