### Zadanie 2 (Watki i Kolejka - Producent-Konsument)
Napisz program o strukturze **producent-consumers**:
1. **Producent**: Generuje kolejne liczby naturalne i dodaje je do kolejki (`queue.Queue`).
2. **Konsument 1**: Pobiera z kolejki tylko liczby **parzyste**.
3. **Konsument 2**: Pobiera z kolejki tylko liczby **nieparzyste**.

Uzyj watkow do realizacji producenta i obu konsumentow. Program powinien zakonczyc sie po przetworzeniu okreslonej puli liczb.

```python
import queue
import threading
import time

N = 20  # ilosc liczb do wygenerowania
q = queue.Queue()
parzyste = []
nieparzyste = []
producer_done = threading.Event()
```

```python
def producer():
    for i in range(1, N + 1):
        q.put(i)
        print(f"  [producent] dodaje {i}")
        time.sleep(0.02)
    producer_done.set()
    print("[producent] zakonczyl prace")


def consumer_even():
    while True:
        try:
            item = q.get(timeout=0.2)
        except queue.Empty:
            if producer_done.is_set() and q.empty():
                break
            continue
        if item % 2 == 0:
            parzyste.append(item)
            print(f"[konsument parzysty]    pobiera {item}")
        else:
            q.put(item)  # oddaje liczbe nieparzysta drugiemu konsumentowi
            time.sleep(0.005)


def consumer_odd():
    while True:
        try:
            item = q.get(timeout=0.2)
        except queue.Empty:
            if producer_done.is_set() and q.empty():
                break
            continue
        if item % 2 == 1:
            nieparzyste.append(item)
            print(f"[konsument nieparzysty] pobiera {item}")
        else:
            q.put(item)
            time.sleep(0.005)
```

```python
watki = [
    threading.Thread(target=producer, name="producent"),
    threading.Thread(target=consumer_even, name="konsument_parzysty"),
    threading.Thread(target=consumer_odd, name="konsument_nieparzysty"),
]

for t in watki:
    t.start()
for t in watki:
    t.join()

print()
print("Parzyste:   ", sorted(parzyste))
print("Nieparzyste:", sorted(nieparzyste))
print(f"Suma:        {len(parzyste) + len(nieparzyste)} (oczekiwane: {N})")
```
